# 02 — Teams & Standings

Explore team objects: rosters, records, standings, and team-level stats.

```python
import sys
sys.path.insert(0, "../src")
from connection import get_league
import pandas as pd

league = get_league()
print(f"Connected: {league.settings.name} ({league.year})")
```

## All Teams Overview

```python
rows = []
for t in league.teams:
    rows.append({
        "ID": t.team_id,
        "Name": t.team_name,
        "Abbrev": t.team_abbrev,
        "W": t.wins,
        "L": t.losses,
        "T": t.ties,
        "PF": round(t.points_for, 1),
        "PA": round(t.points_against, 1),
        "Acquisitions": t.acquisitions,
        "Trades": t.trades,
        "Standing": t.standing,
    })

df = pd.DataFrame(rows).sort_values("Standing")
df
```

## Standings (sorted by league method)

```python
standings = league.standings()
for i, t in enumerate(standings, 1):
    print(f"{i:2d}. {t.team_name:30s}  {t.wins}-{t.losses}-{t.ties}")
```

## Team Object Deep Dive

Inspect all attributes on a single team.

```python
team = league.teams[0]
print(f"Team: {team.team_name}")
print(f"  team_id:        {team.team_id}")
print(f"  team_abbrev:    {team.team_abbrev}")
print(f"  division_id:    {team.division_id}")
print(f"  division_name:  {team.division_name}")
print(f"  standing:       {team.standing}")
print(f"  final_standing: {team.final_standing}")
print(f"  acquisitions:   {team.acquisitions}")
print(f"  drops:          {team.drops}")
print(f"  trades:         {team.trades}")
print(f"  owners:         {team.owners}")
print(f"  logo_url:       {team.logo_url}")
print(f"  roster size:    {len(team.roster)}")
print(f"  schedule size:  {len(team.schedule)}")
```

## Team Stats

```python
print(f"Team stats for {team.team_name}:")
if team.stats:
    for k, v in sorted(team.stats.items()):
        print(f"  {k}: {v}")
else:
    print("  (no team stats available)")
```

## Roster Breakdown

```python
rows = []
for p in team.roster:
    rows.append({
        "Name": p.name,
        "Position": p.position,
        "Lineup Slot": p.lineupSlot,
        "Pro Team": p.proTeam,
        "Injury": p.injuryStatus,
        "Acq Type": p.acquisitionType,
        "Eligible": ", ".join(p.eligibleSlots),
    })

pd.DataFrame(rows)
```

## Team Schedule (matchups this season)

```python
for i, matchup in enumerate(team.schedule[:10], 1):
    opp = matchup.away_team if matchup.home_team == team else matchup.home_team
    opp_name = opp.team_name if hasattr(opp, 'team_name') else str(opp)
    side = "vs" if matchup.home_team == team else "@"
    print(f"  Week {i:2d}: {side} {opp_name:30s}  Score: {matchup.home_final_score}-{matchup.away_final_score}  Winner: {matchup.winner}")
```
