# 01 — League Info

Connect to the ESPN Fantasy Basketball league and explore league-level settings:
scoring type, roster slots, stat categories, schedule, and season metadata.

```python
import sys
sys.path.insert(0, "../src")
from connection import get_league

league = get_league()
print(f"Connected: {league.settings.name} ({league.year})")
```

## League Settings

```python
s = league.settings
print(f"Name:            {s.name}")
print(f"Scoring Type:    {s.scoring_type}")
print(f"Teams:           {s.team_count}")
print(f"Playoff Teams:   {s.playoff_team_count}")
print(f"Reg Season:      {s.reg_season_count} weeks")
print(f"Veto Votes Req:  {s.veto_votes_required}")
print(f"Trade Deadline:  {s.trade_deadline}")

# Inspect all available attributes
print("\n--- All settings attributes ---")
for attr in sorted(dir(s)):
    if not attr.startswith('_'):
        print(f"  {attr}: {getattr(s, attr)}")
```

## Roster Slot Configuration

Which lineup positions are available? How many bench/IR slots?

```python
import pandas as pd

# Roster slots are on the settings object
if hasattr(s, 'roster'):
    print("Roster slots:", s.roster)

# Also inspect a team's roster to see which lineup slots exist
team = league.teams[0]
slot_counts = {}
for p in team.roster:
    slot = p.lineupSlot
    slot_counts[slot] = slot_counts.get(slot, 0) + 1

print(f"\nRoster slots on {team.team_name}:")
for slot, count in sorted(slot_counts.items()):
    print(f"  {slot}: {count}")

# Eligible slots for a sample player
sample = team.roster[0]
print(f"\nSample player: {sample.name}")
print(f"  Position: {sample.position}")
print(f"  Eligible slots: {sample.eligibleSlots}")
```

## Scoring Categories

For H2H Categories leagues, which stats count? Check a box score to see.

```python
# Pull a box score to see which stat categories are used
box_scores = league.box_scores(matchup_period=1)
bs = box_scores[0]

print(f"BoxScore class: {type(bs).__name__}")

if hasattr(bs, 'home_stats'):
    print(f"\nScoring categories ({len(bs.home_stats)}):")
    for cat, data in bs.home_stats.items():
        print(f"  {cat}: {data}")
else:
    print("No home_stats — this may be a points league.")
    print(f"Home score: {bs.home_score}")
```

## Season Metadata

```python
print(f"Current matchup period: {league.currentMatchupPeriod}")
print(f"Scoring period ID:      {league.scoringPeriodId}")
print(f"First scoring period:   {league.firstScoringPeriod}")
print(f"Final scoring period:   {league.finalScoringPeriod}")
print(f"Current week:           {league.current_week}")
print(f"Number of teams:        {len(league.teams)}")
print(f"Previous seasons:       {league.previousSeasons}")
```

## Pro Team Schedule

The league object has a `pro_schedule` dict that maps NBA team IDs to their game schedules.

```python
# Inspect the pro schedule structure
print(f"Pro schedule keys (team IDs): {sorted(league.pro_schedule.keys())}")

# Look at one team's schedule
sample_team_id = list(league.pro_schedule.keys())[0]
sample_sched = league.pro_schedule[sample_team_id]
print(f"\nSample schedule for team ID {sample_team_id}:")
print(f"  Type: {type(sample_sched)}")
# Show first few entries
for i, (k, v) in enumerate(sorted(sample_sched.items())[:5]):
    print(f"  Period {k}: {v}")
```
